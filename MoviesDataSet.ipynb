```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

df = pd.read_csv('/Users/arhamakhtar/Downloads/movies.csv')
```

```python
df
```

```python
# checking for missing data
for col in df.columns:
    missing_data = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing_data*100)))
```

```python
#data types for our columns 
df.dtypes
```

```python
#cleaning data
#change data type
df = df.fillna(0)
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#to display all the data
pd.set_option('display.max_rows', None)
```

```python
#drop any duplicates
#df.drop_duplicates()
```

```python
#checking corelaton between budget and gross revenue
#scatter plot with budget and gross
 
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Gross Earnings')
plt.ylabel('Film Budget') 
plt.show()
```

```python
sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"red"}, line_kws={"color":"green"})
```

```python
df.corr(method="pearson")
```

```python
correlation_matrix = df.corr(method="pearson")
sns.heatmap(correlation_matrix, annot =True)
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

```python
#numeric representation of all the objects
df_num = df
for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name]= df_num[col_name].astype('category')
        df_num[col_name]= df_num[col_name].cat.codes #gives random numbers
        
```

```python
correlation_matrix = df_num.corr(method="pearson")
sns.heatmap(correlation_matrix, annot =True)
plt.title('Correlation Matrix for All Features')
plt.show()
```

```python
df_num.corr()
```

```python
# to check the highest corr
x = df_num.corr()
corr_pairs =x.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
#finding highest correlation
high_corr = sorted_pairs[(sorted_pairs) >= 0.5]
high_corr
```
